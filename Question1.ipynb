```python
marks =[]
for i in range(5):
    mark = float(input(f"Enter mark of student {i+1}: "))
    marks.append(mark)
```

```python
highest = max(marks)
lowest = min(marks)
average = sum(marks) / len(marks)

above_average = 0
for mark in marks:
    if mark > average:
        above_average += 1
```

```python
print("\nResults:")
print("Highest Mark:", highest)
print("Lowest Mark:", lowest)
print("Average Mark:", average)
print("Students Scoring Above Average:", above_average)
```
