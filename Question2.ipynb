```python
employees = {
    "Ravi": 50000,
    "Sita": 60000,
    "Kiran": 45000,
    "Anu": 70000,
    "Rahul": 55000
}
```

```python
highest_employee = max(employees, key=employees.get)
lowest_employee = min(employees, key=employees.get)

average_salary = sum(employees.values()) / len(employees)
```

```python
average_salary = sum(employees.values()) / len(employees)

print("Employee with Highest Salary:")
print(highest_employee, "-", employees[highest_employee])

print("\nEmployee with Lowest Salary:")
print(lowest_employee, "-", employees[lowest_employee])

print("\nAverage Salary:", average_salary)

print("\nEmployees with Salary Greater Than Average:")
for name, salary in employees.items():
    if salary > average_salary:
        print(name, "-", salary)
```
